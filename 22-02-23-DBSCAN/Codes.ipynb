```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples = 500)
plt.scatter(X[:, 0], X[:, 1], c = y)
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters = 3).fit(X)
plt.scatter(X[:, 0], X[:, 1], c = km.predict(X))
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN().fit(X)
plt.scatter(X[:, 0], X[:, 1], c = db.labels_)
```

```python
db.labels_
```

```python
from sklearn.datasets import make_moons

X, y = make_moons(noise = 0.1, n_samples = 500)
plt.scatter(X[:, 0], X[:, 1], c = y)
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps = 0.3, min_samples = 3).fit(X)
plt.scatter(X[:, 0], X[:, 1], c = db.labels_)
```

```python
db.labels_
```

```python
from sklearn.neighbors import NearestNeighbors

neigh = NearestNeighbors(n_neighbors=7)
neigh.fit(X)
distance, index = neigh.kneighbors(X)
```

```python
dis = distance[:, 6].copy()
dis.sort()
plt.plot(dis)
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps = 0.1, min_samples = 6).fit(X)
plt.scatter(X[:, 0], X[:, 1], c = db.labels_)
print(np.unique(db.labels_))
```

## 3 circles

```python
from sklearn.datasets import make_circles
```

```python
def circle(samples = 1500):
    X, y = make_circles(n_samples = samples, noise = 0.1, factor = 0.3)
    X = X[ y==0 ]
    y = y[ y== 0]
    return X, y

X1, y1 = circle(2000)
X2, y2 = circle(1000)
X3, y3 = circle()

X1 *= 10
X2 *= 25
X3 *= 50

X = np.vstack([X1, X2, X3])

plt.scatter(X[:, 0], X[:, 1])
```

```python
x = np.random.rand(100, 2)*120 - 60

plt.scatter(x[:, 0],x[:, 1], color = 'r')
plt.scatter(X[:, 0], X[:, 1])
```

```python
X = np.vstack([X, x])

plt.scatter(X[:, 0], X[:, 1])
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters = 3).fit(X)
plt.scatter(X[:, 0], X[:, 1], c = km.predict(X))
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], color = 'r')
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import DBSCAN

scaled = MinMaxScaler().fit_transform(X)

db = DBSCAN().fit(scaled)
plt.scatter(scaled[:, 0], scaled[:, 1], c = db.labels_)
print(np.unique(db.labels_))
```

```python
from sklearn.neighbors import NearestNeighbors

neigh = NearestNeighbors(n_neighbors=11)
neigh.fit(scaled)
distance, index = neigh.kneighbors(scaled)

dis = distance[:, 6].copy()
dis.sort()
plt.plot(dis)
```

```python
db = DBSCAN(eps = 0.04, min_samples=6).fit(scaled)
plt.scatter(scaled[:, 0], scaled[:, 1], c = db.labels_)
print(np.unique(db.labels_))
```
